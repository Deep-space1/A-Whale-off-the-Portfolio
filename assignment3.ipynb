```python
#Importing Libraries

import pandas as pd
import datetime as dt
from pathlib import Path
import numpy as np
%matplotlib inline
import seaborn as sns
```

```python
#Accessing data from csv files
Algo_data = Path("algo_returns.csv")
Otex_data = Path("otex_historical.csv")
sp_tsx_data = Path("sp_tsx_history.csv")
l_historical_data = Path("l_historical.csv")
shop_historical_data= Path("shop_historical.csv")
whale_data = Path("whale_returns.csv")

df_algo = pd.read_csv(Algo_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_otex = pd.read_csv(Otex_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_sp_tsx = pd.read_csv(sp_tsx_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_l_historical = pd.read_csv(l_historical_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_shop = pd.read_csv(shop_historical_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_whale = pd.read_csv(whale_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
```

### S&P TSX 60 Returns

```python
# Checking null values in TSX 60 returns

df_sp_tsx.isnull().sum()
```

```python
# Accessing data

df_sp_tsx.head()
```

```python
# Formating data

df_sp_tsx['Close'] = df_sp_tsx['Close'].str.replace("$","")
df_sp_tsx['Close'] = df_sp_tsx['Close'].str.replace(",","")

df_sp_tsx.head()
```

### Combined Returns

```python
# Merging all the portfolio.

combined_returns = pd.concat([df_whale, df_algo, df_sp_tsx], axis = 'columns', join = 'inner')
combined_returns
```

```python
# Plotting daily returns for all the portfolios

combined_returns.plot(figsize=(10,10), title='Daily Returns')
```

```python
# Calculating Cumullative Returns for all portfolios.

combined_cummulative_returns = (1 + combined_returns).cumprod()
combined_cummulative_returns
```

```python
# Visual plotting to compare portfolios with each other

combined_cummulative_returns.plot(figsize=(10,10), title='Cumulative Returns')
```

```python
# Boxplotting visual for risk assessment.

combined_returns.boxplot(fontsize=10, figsize=(20,10))
```

```python
# Determining the least volatile portfolio.

combined_std = combined_returns.std()
combined_std.sort_values(ascending=False, inplace=True)
combined_std
```

```python
# Calculating annualized standard deviation for each portfolio.

combined_annual_std = combined_std * np.sqrt(252)
combined_annual_std.sort_values(ascending=False, inplace=True)
combined_annual_std
```

```python
# Calculating Rolling Covariance of a selected portfolio with a 60-Day window

rolling_covariance = combined_returns['Algo 1'].rolling(window=60).cov()
```

```python
# Calculating Rolling variance of a S&P TSX 60 Returns portfolio with a 60-Day window

rolling_variance = combined_returns['S&P_TSX_60_Returns'].rolling(window=60).var()
```

```python
# Calculatting Beta of Algo 1 comparted to S&P TSX 60 Returns.


algo1_beta = rolling_covariance / rolling_variance
algo1_beta.dropna(inplace=True)
algo1_beta
```

```python
# Plotting Algo 1 Beta.

algo1_beta.plot(figsize=(10,10), title='Rolling 60 Day Beta for Algo1' )
```

```python
# Using .ewm to calculate exponential weighted moving average with 21-Day halflife window.

df_ewm = combined_returns.ewm(halflife=21, adjust=True).mean()
df_ewm
```

### Sharpe Ratios

```python
# Annualized Sharpe Ratios.

df_sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
df_sharpe_ratios
```

```python
# Visual plot of sharpe ratios

df_sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

## Custom Portfolio

```python
# Accessing data from csv files.
# Resetting date index and inferring datetime format.

tsla_data = Path("TSLA.csv")
fb_data = Path("FB.csv")
shop_data = Path("SHOP.csv")

df_tsla = pd.read_csv(tsla_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_fb = pd.read_csv(fb_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
df_shop = pd.read_csv(shop_data, index_col="Date", infer_datetime_format=True, parse_dates=True)

```

```python
# Formatting data by converting closing prices into daily returns.
# Dropping Null values
# Assigning ticker symbols to column names.


df_tsla = df_tsla.pct_change()
df_tsla.dropna(inplace=True)
df_tsla = df_tsla.rename(columns={
    "Close" : "TSLA_returns"
}) 

df_fb = df_fb.pct_change()
df_fb.dropna(inplace=True)
df_fb = df_fb.rename(columns={
    "Close" : "FB_returns"
}) 

df_shop = df_shop.pct_change()
df_shop.dropna(inplace=True)
df_shop = df_shop.rename(columns={
    "Close" : "SHOPIFY_returns"
}) 
```

```python
# Merging all Portfolios together.

cus_returns = pd.concat([df_tsla, df_fb, df_shop], axis = 'columns', join = 'inner')
cus_returns.head()
```

```python
#Calculating Weighted Returns

weights = [1/3,1/3,1/3]
weight_returns = cus_returns.dot(weights)


weight_returns = weight_returns.rename("Personal_Returns")
weight_returns.head()  
```

```python
# Merging custom portfolio to original dataframe.

all_returns = pd.concat([df_whale, df_algo, df_sp_tsx, weight_returns], axis = 'columns', join = 'inner')
all_returns
```

```python
# Calculating standard deviation to compare custom portfolio with others


all_std = all_returns.std()
all_std.sort_values(ascending=False, inplace=True)
all_std
```

```python
# Determining  annual standard deviation

all_annual_std = all_std * np.sqrt(252)
all_annual_std.sort_values(ascending=False, inplace=True)
all_annual_std
```

```python
# Rolling Window Standard Deviation with 21-Day window.

rolling_all = all_returns.rolling(window=21).std().dropna().copy()
rolling_all
```

```python
# Plotting rolling STD.

rolling_all.plot(figsize=(15,15))
```

```python
# Determing Correlation of Personal returns with others.

all_correlation = all_returns.corr()
all_correlation
```

```python
# Plotting Correlation

sns.heatmap(all_correlation, annot=True)
```

```python
# Analysinng covariance of Personal Returns with a 60-Day rolling window

rolling_covariance_all = all_returns['Personal_Returns'].rolling(window=60).cov()
rolling_covariance_all.dropna(inplace=True)
rolling_covariance_all
```

```python
# Calculatting variance of S&P TSX 60 Returns with a 60-Day rolling window.

rolling_variance_all = all_returns['S&P_TSX_60_Returns'].rolling(window=60).var()
rolling_variance_all.dropna(inplace=True)
rolling_variance_all
```

```python
# Determining Beta of Personal Returns compared to S&P TSX 60 Returns with a 60-Day rolling window.

personal_beta = rolling_covariance_all / rolling_variance_all
personal_beta.dropna(inplace=True)
personal_beta
```

```python
# Plotting Personal Returns Beta with a 60-Day window

personal_beta.plot(figsize=(10,10), title='Rolling 60 Day Beta for Personal_Returns against TSX 60' )
```

```python
# Calculating Annualized Sharpe Ratios 

df_sharpe_ratios_all = (all_returns.mean() * 252) / (all_returns.std() * np.sqrt(252))
df_sharpe_ratios_all
```

```python
# Plotting Sharpe Ratios

df_sharpe_ratios_all.plot(kind='bar', title="Custom Annual Sharpe Ratio")
```
